```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(context="paper", style="ticks", font_scale=1.5)
df = pd.read_csv("../out/f5eaa9b8-16e/train_loss.csv", names=["wall time", "loss"])
df["wall time"] = df["wall time"] - df["wall time"].min()
df["loss (moving average)"] = df["loss"].rolling(7).mean()

fig, ax = plt.subplots(figsize=(6, 4))
sns.lineplot(data=df, x="wall time", y="loss (moving average)", ax=ax)
```
