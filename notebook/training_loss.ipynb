```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(context="paper", style="ticks", font_scale=2)
df = pd.read_csv("../out/81392c79-89e/scalar.csv", names=["wall time (sec)", "type", "value"])
df["type"] = df["type"].replace("manifold classification loss", "discriminator loss")


start_timestamp = df["wall time (sec)"].min()

df["wall time (sec)"] = df["wall time (sec)"] - start_timestamp

g = sns.FacetGrid(df, row="type", aspect=3, sharey=False)
g.map(sns.scatterplot, "wall time (sec)", "value")
#fig, ax = plt.subplots(figsize=(10, 8))
#p_loss = sns.lineplot(data=df, x="wall time", y="value", hue="type")
g.set_titles("{row_name}")
g.axes[0, 0].set_ylim((0, 0.75))
g.axes[1, 0].set_ylim((0, 0.75))
g.axes[2, 0].set_ylim((0.84, 0.91))
g.set_ylabels("")
plt.savefig("sup1.png", dpi=200)
```
