```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
fit, axes = plt.subplots(figsize=(9, 6), nrows=2, ncols=2, constrained_layout=True)

df = pd.read_csv(os.path.join("res", "tuning", "yahoo_answer_adamix_2000.csv"),
                 names=["coefficient intrusion", "mixup layer", "best valid accuracy"],
                 header=0)

sns.set_theme(context="paper", style="ticks", font_scale=1.5)
sns.lineplot(x="coefficient intrusion", y="best valid accuracy", data=df, ax=axes[0, 0])
sns.set_theme(context="paper", style="ticks", font_scale=1.5)
sns.lineplot(x="mixup layer", y="best valid accuracy", data=df, ax=axes[0, 1])

plt.show()
```
